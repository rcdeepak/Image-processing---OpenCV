### 1. Image Thresholding with openCV

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('Images/image.jpg')
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
#Creates a binary image from a gray image based on the threshold values.
ret, thresh1 = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap='gray')
```

```python
#Binary Inverse
ret, thresh2= cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh2, cmap='gray')
```

### 2. Contours Detection with OpenCV

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img= cv2.imread('Images/hand.jpg')
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
ret, thresh = cv2.threshold(img_gray, 127, 255, 0)
contours, hierarchy=cv2.findContours(thresh, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(img_gray, contours, -1, (0, 0, 0), 3)
plt.figure(figsize=(6,6))
plt.imshow(img_gray, cmap='gray')

cv2.waitKey(0)
cv2.destroyAllWindows()
```
